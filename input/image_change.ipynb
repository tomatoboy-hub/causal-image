```python
import os
import shutil

def rename_and_move_images(src_folder, dst_folder):
    # 移動先のフォルダが無ければ作成
    os.makedirs(dst_folder, exist_ok=True)
    
    # src_folder 内のすべてのアイテムを処理
    for filename in os.listdir(src_folder):
        old_path = os.path.join(src_folder, filename)

        # ファイルではない(=ディレクトリ等)はスキップ
        if not os.path.isfile(old_path):
            continue
        
        # 既に "watch" で始まる場合はそのまま、それ以外は "watch" を付与
        if filename.startswith("cloth"):
            new_filename = filename
        else:
            new_filename = "cloth" + filename

        # 実際に移動・リネーム先となるパス
        new_path = os.path.join(dst_folder, new_filename)

        # shutil.move でファイルを移動 + リネーム
        shutil.move(old_path, new_path)
        print(f"Moved: {old_path} -> {new_path}")


if __name__ == "__main__":
    rename_and_move_images("cloth_imgs", "shoclo_imgs")

```
